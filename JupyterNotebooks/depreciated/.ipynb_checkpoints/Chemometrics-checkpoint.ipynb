```python
import numpy as np
import pandas as pd
import sklearn.preprocessing as pre
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression as PLS
import loadTestData as load_data # Imported from accompanying library (loadTestData.py)
import supportMethods as supp
```

```python
# Parameters
##### PCA
num_proj_plots = 3

##### PLS
num_classes = 2
```

```python
# Load data
path_to_data = 'C:\\Users\\csa97\\Research\\Projects\\DuLab\\ADAP-ML\\adap-ml\\data\\SCLC_study_output_filtered_2.csv'
df = pd.read_csv(path_to_data, index_col=0)
sample_names = df.index.values.tolist()
variable_names = df.columns.values.tolist()
M = len(sample_names)
N = len(variable_names)
```

```python
# Normalize data: Auto-Scaling
#     - Mean Center
#     - Uniform Variance scaling
scaler = pre.StandardScaler().fit(df)
df_norm = scaler.transform(df)
```

```python
# Perform PCA
pca = PCA().fit(df_norm)
pc = pca.transform(df_norm)
```

```python
# Perfrom PLA-DA
#### vector of class responses associated with data
resp = load_data.getResponseMatrix2D()

pls = PLS().fit(df_norm, resp)
pls_trans = pls.transform(df_norm)
```

```python
#Visualize PCA Results
#### Get and plot scree and cumulative variance of components
supp.plotScree(pca.explained_variance_, len(pca.explained_variance_))

#### Plot PC feature projections
supp.plotProjectionScatterMultiClass(pc, resp, num_proj_plots)

#### Plot Eigenvector loadings
varis = df.columns.values.tolist()
supp.plotVectorLoadings(pca.components_, varis, num_proj_plots)

#### Plot Eigenvector Loading Scatter Plots
#plotLoadingsScatter(pca.components_)
```

```python
# Visualize PLS-Results
supp.plotProjectionScatterMultiClass(pls_trans, resp, num_classes)
```
